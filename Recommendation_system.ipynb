```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
stopwords_list = stopwords.words('english')+ stopwords.words('portuguese')
```

```python
vectorizer = TfidfVectorizer(analyzer='word',
                max_features=5000,
                stop_words=stopwords_list
)
```

```python
tfidf_matrix = vectorizer.fit_transform(shared_articles['text'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.model_selection import train_test_split
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

# drop when personId, contentId are the same.
users_interactions_unique = users_interactions.drop_duplicates(subset=['personId', 'contentId'])

# Create the pivot table
user_item_matrix = users_interactions_unique.pivot(index='personId', columns='contentId', values='rating').fillna(0)

# Convert it to a sparse matrix
user_item_sparse = csr_matrix(user_item_matrix.values)

# Fit to find similar users
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(user_item_sparse)
```

```python
# Finding recommendations based on user interaction
def recommend_for_user(user_id):
    user_index = user_item_matrix.index.get_loc(user_id)
    distances, indices = knn.kneighbors(user_item_sparse[user_index], n_neighbors=6)

    for i in range(1, len(distances.flatten())):
        print(f"Recommendation {i}: {user_item_matrix.columns[indices.flatten()[i]]}")
```

```python
recommend_for_user(-9223121837663643404)
```
